```python
def create_dev_position(df):
    """
    Create a new column for the department x y for chart.
    """

    dev_x_dict = {}
    # for now
    df['Development_Stage'].fillna('Deployment', inplace=True)
    
    devs = df['Development_Stage'].unique().tolist()
    devs.sort()
    # 5 wide means first is at 3 then +5
    i = 0
    for dev in devs:
        dev_x_dict[dev] = i
        i += 1

    df['dev_rn'] = df.sort_values(['Development_Stage','Title'], ascending=[True,True]) \
             .groupby(['Development_Stage']) \
             .cumcount()
    df['dev_x'] = (df['Development_Stage'].map(dev_x_dict) * 7) + (df['dev_rn'] % 5) +1
    df['dev_y'] = df['dev_rn'] // 5 + 1

    return df
    
```

```python
def create_dept_position(df):
    """
    Create a new column for the department x y for chart.
    """

    dept_x_dict = {}
    depts = df['Department_Code'].unique().tolist()
    
    depts.sort()
    # 5 wide means first is at 3 then +5
    i = 0
    for dept in depts:
        dept_x_dict[dept] = i
        i += 1

    df['dept_rn'] = df.sort_values(['Department_Code','Title'], ascending=[True,True]) \
             .groupby(['Department_Code']) \
             .cumcount()
    df['dept_x'] = (df['Department_Code'].map(dept_x_dict) * 7) + (df['dept_rn'] % 5) +1
    df['dept_y'] = df['dept_rn'] // 5 + 1

    return df
    
```

```python
data = create_dept_position(data)
data = create_dev_position(data)
```
